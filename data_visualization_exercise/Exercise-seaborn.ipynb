### 使用資料集--Kaggle酪梨價格資料集

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
avocado=pd.read_csv('./avocado.csv')
```

```python
avocado.drop(columns='Unnamed: 0',inplace=True)
```

```python
print(avocado.head())
print(avocado.isnull().sum())
```

# Scatter plot--散點圖



## sns.scatterplot(x=None,y=None,hue=None,data=None,legend='brief',ax=None)



```python
z0=plt.figure(figsize=(10,10))
sns.scatterplot(x='AveragePrice',y='Total Bags',data=avocado)
plt.title('AveragePrice and Total Bags scatter plot')
plt.show()
```

```python
# highlight by type
z1=plt.figure(figsize=(10,10))
sns.scatterplot(x='AveragePrice',y='Total Bags',hue='type',data=avocado)
plt.title('AveragePrice and Total Bags scatter plot by type')
plt.show()
```

# Line chart--折線圖

## sns.lineplot(x=None,y=None,hue=None,data=None)

```python
# average price 折線圖
z2=plt.figure(figsize=(10,10))
sns.lineplot(x='year',y='AveragePrice',data=avocado)
plt.show()
```

```python
# 使用type欄位分組
z3=plt.figure(figsize=(10,10))
sns.lineplot(x='year',y='AveragePrice',hue='type',style='type',markers=True,data=avocado)
plt.show()
```

# Histogram--直方圖


scipy.stats.norm--常態分配


使用子模組使常態分配出現在圖上

## sns.distplot(a,bins=None,hist=True,kde=True,normhist=False,fit=scipy.stats.norm)

```python
from scipy.stats import norm
z4=plt.figure(figsize=(10,10))
sns.distplot(avocado['AveragePrice'],bins=50,fit=norm)
plt.title('average price histogram')
plt.show()
```

```python
z5=plt.figure(figsize=(10,10))
sns.distplot(avocado['AveragePrice'],bins=10,fit=norm)
plt.title('average price histogram')
plt.show()
```

# bar chart--長條圖

## sns.countplot(x=None,y=None,hue=None,data=None)

```python
sns.countplot(x='type',data=avocado)
```

```python
# 使用年分組
z6=plt.figure(figsize=(15,10))
sns.countplot(x='type',hue='year',data=avocado)
plt.show()
```

# Box plot--箱線圖

## sns.boxplot(x=None,y=None,hue=None,notch=False)

```python
z7=plt.figure(figsize=(10,10))
sns.boxplot(x='type',y='AveragePrice',hue='year',data=avocado)
plt.show()
```

# Heatmap--熱圖

## sns.heatmap(data=None,annot=None,cbar=True)

```python
correlation=avocado.drop(columns='year').corr()
print(correlation)
```

```python
z8=plt.figure(figsize=(10,10))
sns.heatmap(correlation,annot=True)
plt.title('heatmap for avocado dataset')
plt.show()
```

# pairplot--多變量圖

兩個特徵互相比較

## sns.pairplot(data=None,hue=None,kind='scatter',dropna=True)

```python
t1=plt.figure(figsize=(10,10))
sns.pairplot(avocado.drop(columns='year'))
plt.show()
```

```python
# 使用類別作為顏色區別
t2=plt.figure(figsize=(10,10))
sns.pairplot(avocado.drop(columns='year'),hue='type')
plt.show()
```
